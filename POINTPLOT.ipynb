1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import sklearn
````

````python

df = pd.read_csv('Profissão- Cientista de Dados_M10_support material.csv')
````

````python
# As pessoas(5833 pessoas) que possuem veículo têm renda maior que 4890

# As pessoas(9167 pessoas) que não possuem veículo têm renda maior que 4300 e menor ou igual 4500

plt.style.use('ggplot')

sns.pointplot(x='posse_de_veiculo', y ='renda', data = df, dodge=True, ci = 95).set(title='Posse de posse_de_veículo x Renda');
````

````python
# As pessoas(10082 pessoas) que possuem imóvel têm renda maior que 4600

# As pessoas(4918 pessoas) que não possuem imóvel têm renda maior ou igual 4300 e menor que 4600

sns.pointplot(x='posse_de_imovel', y ='renda', data = df, dodge=True, ci = 95)

plt.title('Posse de Imóvel x Renda');
````

````python

sns.pointplot(x='posse_de_veiculo', y = 'renda', hue ='posse_de_imovel', data = df,dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 2);
````

````python

# Variaveis --> 'posse_de_imovel','posse_de_veiculo','renda'

# Calculando o percental das variaveis 'posse_de_imovel','posse_de_veiculo' em função da renda

df_previsao_renda = df[['posse_de_imovel','posse_de_veiculo','renda']].copy()

df_previsao_renda['percent'] = (df_previsao_renda['posse_de_imovel'] | df_previsao_renda['posse_de_veiculo']) / (
    df_previsao_renda['renda'] - (df_previsao_renda['posse_de_imovel'] | df_previsao_renda['posse_de_veiculo']))

sns.pointplot(x='posse_de_veiculo', y = 'percent', hue ='posse_de_imovel', data = df_previsao_renda,dodge=True, ci = 95);

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 2);
````
